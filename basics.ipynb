```python
no_students = int(input("Enter the number of students:"))
list_students = []
for i in range(1, no_students+1):
    name = input("name:")
    rollno = int(input("roll number:"))
    marks = int(input("marks:"))

    list_students.append({
        'name': name,
        'rollno': rollno,
        'marks': marks
    })


print(list_students)
```
