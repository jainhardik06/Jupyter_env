```python
import numpy as np
import pandas as pd
import matplotlib.pylab as plt
import cv2 as cv
import os
from skimage import io
from PIL import Image
```

```python
img = cv.imread('tajmahal.jpg')
plt.imshow(img)
```

```python
img_rgb=cv.cvtColor(img,cv.COLOR_BGR2RGB)
plt.figure(figsize=(18,6))
plt.subplot(141)
plt.title('Original Image')
plt.imshow(img_rgb)
plt.axis('off')

plt.subplot(142)
plt.title('Mirror Image on x axis')
plt.imshow(img_rgb[:,::-1])
plt.axis('off')

plt.subplot(143)
plt.title('Mirror Image on y axis')
plt.imshow(img_rgb[::-1,:])
plt.axis('off')

plt.subplot(144)
plt.title('Mirror Image on x and y axis')
plt.imshow(img_rgb[::-1,::-1])
plt.axis('off')
```

```python
if img is None:
    print("Could not read the image.")
else:
    img_fh=cv.flip(img_rgb,0)
    img_fv=cv.flip(img_rgb,1)
    img_fhv=cv.flip(img_rgb,-1)

    plt.figure(figsize=(18,6))
    plt.subplot(131)
    plt.title('Flip Horizontal')
    plt.imshow(img_fh)
    plt.axis('off')

    plt.subplot(132)
    plt.title('Flip Vertical')
    plt.imshow(img_fv)
    plt.axis('off')

    plt.subplot(133)
    plt.title('Flip Horizontal and Vertical')
    plt.imshow(img_fhv)
    plt.axis('off')
    plt.show()
    
```

```python
img1=np.zeros((256,256,3),dtype=np.uint8)

img1 = cv.line(img1,(255,0),(55,255),(255,255,0),8)
plt.imshow(img1)
plt.title('Line')
plt.axis('on')
plt.show()
```
