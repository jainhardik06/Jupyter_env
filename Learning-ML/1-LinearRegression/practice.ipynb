```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import linear_model
```

```python
df = pd.read_csv('canada_per_capita_income.csv')
df.head(3)
```

```python
plt.xlabel('year')
plt.ylabel('per capita income (USD)')
plt.scatter(df['year'], df['per capita income (US$)'], color = 'red', marker='+' )
```

```python
xdf = df.drop('year', axis='columns')
ydf = df.year
reg = linear_model.LinearRegression()
reg.fit(xdf,ydf)
```

```python
reg.predict([[2020]])
```

```python
plt.xlabel('per capita income (US$)')
plt.ylabel('year')
plt.scatter(xdf, ydf, color='red', marker='+')
plt.plot(xdf, reg.predict(xdf), color='blue', linewidth=2)
```
