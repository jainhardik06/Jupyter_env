```python
lst = [1, 2, 3, 4, 5]
print("List:", lst)
print("Indexing - lst[0]:", lst[0])
print("Indexing - lst[-1]:", lst[-1])
print("Slicing - lst[1:4]:", lst[1:4])
print("Slicing - lst[::2]:", lst[::2])

lst.append(6)
print("After append:", lst)

lst.extend([7, 8])
print("After extend:", lst)

lst.insert(0, 0)
print("After insert:", lst)

lst.remove(4)
print("After remove:", lst)

popped = lst.pop()
print("After pop:", lst, "Popped:", popped)

print("Length:", len(lst))
print("Index of 5:", lst.index(5))

lst_copy = lst.copy()
print("Copy:", lst_copy)

lst2 = [5, 2, 8, 1, 9]
lst2.sort()
print("After sort:", lst2)

lst3 = [1, 2, 3, 4]
lst3.reverse()
print("After reverse:", lst3)

print("Count of 2:", lst.count(2))

lst4 = [1, 2, 3]
lst4.clear()
print("After clear:", lst4)
```

```python
tpl = (1, 2, 3, 4, 5)
print("Tuple:", tpl)
print("Indexing - tpl[0]:", tpl[0])
print("Indexing - tpl[-1]:", tpl[-1])
print("Slicing - tpl[1:4]:", tpl[1:4])
print("Slicing - tpl[::2]:", tpl[::2])
print("Length:", len(tpl))
print("Count of 2:", tpl.count(2))
print("Index of 3:", tpl.index(3))

tpl2 = (10, 20)
combined = tpl + tpl2
print("Concatenation:", combined)

repeated = tpl * 2
print("Repetition:", repeated)
```

```python
num = 153
sum_of_cubes = 0
temp = num

for digit in str(num):
    sum_of_cubes += int(digit) ** 3

if sum_of_cubes == num:
    print(f"{num} is Armstrong number")
else:
    print(f"{num} is not Armstrong number")

print(f"Sum of cubes: {sum_of_cubes}")
```

```python
num = 121
str_num = str(num)
reversed_num = str_num[::-1]

if str_num == reversed_num:
    print(f"{num} is a palindrome")
else:
    print(f"{num} is not a palindrome")

print(f"Original: {str_num}, Reversed: {reversed_num}")
```

```python
num = 17
is_prime = True

if num < 2:
    is_prime = False
else:
    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            is_prime = False
            break

if is_prime:
    print(f"{num} is a prime number")
else:
    print(f"{num} is not a prime number")
```
