```python
# 1. Maximum of three numbers
nums = [float(input("Enter first number: ")), float(input("Enter second number: ")), float(input("Enter third number: "))]
print("Maximum:", max(nums))
```

```python
# 2. Simple and compound interest
principal = float(input("Enter principal: "))
rate = float(input("Enter annual rate (in %): ")) / 100
time = float(input("Enter time (in years): "))

simple_interest = principal * rate * time
compound_amount = principal * (1 + rate) ** time
compound_interest = compound_amount - principal

print("Simple Interest:", simple_interest)
print("Compound Interest:", compound_interest)
```

```python
# 3. Check if a number is even
n = int(input("Enter a number: "))
if n % 2 == 0:
    print("Even")
else:
    print("Odd")
```

```python
# 4. Decimal equivalents of fractions 1/2 through 1/10
for d in range(2, 11):
    print(f"1/{d} =", 1 / d)
```

```python
# 5. Loop over a sequence
sequence = ["apple", "banana", "cherry"]
for item in sequence:
    print(item)
```

```python
# 6. Countdown using while loop
start = int(input("Enter a starting number: "))
while start >= 0:
    print(start)
    start -= 1
```
